```python
import os
import torch
```

```python
def get_device():
  return 'cuda' if torch.cuda.is_available() else 'cpu'
device = get_device()
print(f'DEVICE: {device}')
```

```python
project_name = 'mlhw6-gan-anime-face'
data_path = './faces/faces'
results_folder = './results'
#ckpt = f"{results_folder}/model-{int(trainer.train_num_steps/trainer.save_and_sample_every)}.pt"
!mkdir -p "{project_name}" "{results_folder}"
```

```python
  !stylegan2_pytorch --data "{data_path}" --name "{project_name}" --results_dir "{results_folder}" --models_dir "{results_folder}" --attn-layers 1 --image-size 64 --batch-size 16 --num-train-steps 38000 --calculate-fid-every 1000
```

```python
model_step = 25
ckpt = f"{results_folder}/{project_name}/model_{model_step}.pt"
!rm -rf "{results_folder}/default/"
!mkdir -p "{results_folder}/default/"
!cp "{ckpt}" "{results_folder}/default/"
```

```python
#!stylegan2_pytorch  --generate
for i in range(1000):
    !stylegan2_pytorch --generate --trunc-psi 0.65 --results_dir "{results_folder}"  --models_dir "{results_folder}" --attn-layers 1 --image-size 64 --num_image_tiles 1 --load-from "{model_step}"
```
